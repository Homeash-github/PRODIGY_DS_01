```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pycountry
```

```python
data = pd.read_csv('/content/World POP.csv')

countries = [country.name for country in pycountry.countries]

filtered_data = data[data['Country Name'].isin(countries)]
```

```python
import matplotlib.cm as cm
cmap=cm.tab10
top10 = filtered_data.sort_values(by='2023', ascending=False).head(10)

plt.figure(figsize=(10,6))
bars = plt.bar(top10['Country Name'], top10['2023'], color=cmap(np.arange(len(top10))))
plt.xlabel('Country')
plt.ylabel('Total Population (in Billion)')
plt.title('Top 10 Countries with Largest Total Population')
plt.xticks(rotation=60, ha='right')

for bar in bars:
    yval = bar.get_height() / 1_000_000_000
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), "{:.2f}B".format(yval), va='bottom', ha='center')

plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np
bottom10 = filtered_data.sort_values(by='2023').head(10)

plt.figure(figsize=(10,6))
bars = plt.bar(bottom10['Country Name'], bottom10['2023'], color=cmap(np.arange(len(bottom10))))
plt.xlabel('Country')
plt.ylabel('Total Population')
plt.title('Top 10 Countries with Least Total Population')
plt.xticks(rotation=60, ha='right')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), "{:,.0f}".format(yval), va='bottom', ha='center')

plt.tight_layout()
plt.show()
```

```python

india_data = filtered_data[filtered_data['Country Name'] == 'India'].iloc[:, 4:-1].T
china_data = filtered_data[filtered_data['Country Name'] == 'China'].iloc[:, 4:-1].T

decade_starts = np.arange(1973, 2024, 10)
india_decade_averages = [india_data.loc[str(start):str(start+9)].mean().values[0] for start in decade_starts]
china_decade_averages = [china_data.loc[str(start):str(start+9)].mean().values[0] for start in decade_starts]

decade_labels = [f"{start}-{start+9}" for start in decade_starts]
plt.figure(figsize=(10, 6))
bar_width = 0.35

india_bars = plt.bar(np.arange(len(decade_labels)), np.array(india_decade_averages)/1000000000, bar_width, label='India')
china_bars = plt.bar(np.arange(len(decade_labels)) + bar_width, np.array(china_decade_averages)/1000000000, bar_width, label='China')

plt.xlabel('Decade')
plt.ylabel('Average Population (in Billions)')
plt.title('Decade-wise Average Population Comparison: India vs China')
plt.xticks(np.arange(len(decade_labels)) + bar_width/2, decade_labels)
plt.legend()

for bar in india_bars + china_bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, "{:.2f}B".format(yval), va='bottom', ha='center')

plt.tight_layout()
plt.show()
```

```python
russia_data = filtered_data[filtered_data['Country Name'] == 'Russian Federation'].iloc[:, 4:-1].T
bangladesh_data = filtered_data[filtered_data['Country Name'] == 'Bangladesh'].iloc[:, 4:-1].T

decade_starts = np.arange(1973, 2024, 10)
russia_decade_averages = [russia_data.loc[str(start):str(start+9)].mean().values[0] for start in decade_starts]
bangladesh_decade_averages = [bangladesh_data.loc[str(start):str(start+9)].mean().values[0] for start in decade_starts]

decade_labels = [f"{start}-{start+9}" for start in decade_starts]
plt.figure(figsize=(10, 6))
bar_width = 0.35

russia_bars = plt.bar(np.arange(len(decade_labels)), np.array(russia_decade_averages)/1000000, bar_width, label='Russian Federation',color="red")
bangladesh_bars = plt.bar(np.arange(len(decade_labels)) + bar_width, np.array(bangladesh_decade_averages)/1000000, bar_width, label='Bangladesh',color="green")

plt.xlabel('Decade')
plt.ylabel('Average Population (in Millions)')
plt.title('Decade-wise Average Population Comparison: Russian Federation vs Bangladesh')
plt.xticks(np.arange(len(decade_labels)) + bar_width/2, decade_labels)
plt.legend()

for bar in russia_bars + bangladesh_bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, "{:.2f}M".format(yval), va='bottom', ha='center')

plt.tight_layout()
plt.show()
```

```python
india_data_2000 = india_data.loc['2010':]
china_data_2000 = china_data.loc['2010':]

plt.figure(figsize=(12, 8))
plt.plot(india_data_2000.index, india_data_2000.values / 1_000_000_000, label='India', marker='o')
plt.plot(china_data_2000.index, china_data_2000.values / 1_000_000_000, label='China', marker='x')

plt.xlabel('Year')
plt.ylabel('Population (in Billions)')
plt.title('Population Growth: India vs China (2010 Onwards)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

years = np.array(range(2021, 2024)).reshape(-1, 1)

india_pop = india_data.loc['2020':].values.astype(int).reshape(-1, 1) / 1_000_000_000
china_pop = china_data.loc['2020':].values.astype(int).reshape(-1, 1) / 1_000_000_000

bangladesh_data = filtered_data[filtered_data['Country Name'] == 'Bangladesh'].iloc[:, 4:-1].T
bangladesh_pop = bangladesh_data.loc['2020':].values.astype(int).reshape(-1, 1) / 1_000_000_000

india_model = LinearRegression().fit(years, india_pop)
china_model = LinearRegression().fit(years, china_pop)
bangladesh_model = LinearRegression().fit(years, bangladesh_pop)

future_years = np.array(range(2020, 2051)).reshape(-1, 1)
india_pred = india_model.predict(future_years)
china_pred = china_model.predict(future_years)
bangladesh_pred = bangladesh_model.predict(future_years)

plt.figure(figsize=(12, 6))
plt.plot(future_years, india_pred, label='India', marker='o')
plt.plot(future_years, china_pred, label='China', marker='x')
plt.plot(future_years, bangladesh_pred, label='Bangladesh', marker='s')

plt.xlabel('Year')
plt.ylabel('Population (in Billions)')
plt.title('Population Projection: India, China, and Bangladesh (2020-2050)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
